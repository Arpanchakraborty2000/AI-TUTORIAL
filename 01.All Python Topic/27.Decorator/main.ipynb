```python
def main_welcome(func):
   ## msg="welcome"
    def sub_welcome_method():
        print("Welcome to the advance python course ")
        func("Welcome every one to this tutorials ")
        print("please learn these concept properly")
        
    return sub_welcome_method()


main_welcome(print)
```

```python
## decorator 
def log_arguments(func):
    def wrapper(*args, **kwargs):
        print(f"ðŸ“˜ Arguments: {args}, {kwargs}")
        result = func(*args, **kwargs)
        print("âœ… Function executed successfully.")
        return result
    return wrapper

@log_arguments
def add(a, b):
    return a + b

print("Result:", add(5, 10))
```
