```python
## private variable always declear with this sysntax (ex: __name )
class Person:
    def __init__(self,name,age,gender):
        self.__name=name ## private variables
        self.__age=age ## private variables
        self.gender=gender  ## public variables
        
        
def get_name(person):
    return person.__name
## Private variable always being used inside the class .. outside the class is not work ..

person =Person("krish",34,"Male")
get_name(person)
```

```python
## Protected variable syntax is (ex : _name)

class Person:
    def __init__(self,name,age,gender):
        self._name=name ## Protected variables
        self._age=age ## Protected variables
        self.gender=gender  ## public variables
        
        
class Employe(Person):
    def __init__(self, name ,age ,gender):
        super().__init__(name,age,gender)
        
        
employee = Employe("Arpan",25,"Male")
print(employee._name)
```

```python
class Person:
    def __init__(self, name, age):
        self._name = name       # Protected variable
        self._age = age         # Protected variable

    # Getter method for name
    def get_name(self):
        return self._name

    # Setter method for name
    def set_name(self, name):
        if len(name) > 1:
            self._name = name
        else:
            print("❌ Name must have at least 2 characters.")

    # Getter method for age
    def get_age(self):
        return self._age

    # Setter method for age
    def set_age(self, age):
        if age > 0:
            self._age = age
        else:
            print("❌ Age must be positive.")
            

# --- Usage ---
person1 = Person("Arpan", 25)

print("Name:", person1.get_name())  # Using getter
print("Age:", person1.get_age())

person1.set_name("Riya")            # Using setter
person1.set_age(30)

print("\nAfter update:")
print("Name:", person1.get_name())
print("Age:", person1.get_age())
```

```python
## Using @property

class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if len(value) > 1:
            self._name = value
        else:
            print("❌ Name must have at least 2 characters.")

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value > 0:
            self._age = value
        else:
            print("❌ Age must be positive.")
            

# --- Usage ---
person = Person("Arpan", 25)

print(person.name)   # Getter used automatically
print(person.age)

person.name = "Riya"  # Setter used automatically
person.age = 30

print(person.name)
print(person.age)
```
