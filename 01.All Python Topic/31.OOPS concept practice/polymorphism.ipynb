```python
## Method Overriding


class Car:
    def drive(self):
        print("Car is driving with fuel engine")

class ElectricCar(Car):
    def drive(self):   # overriding parent method
        print("Car is driving with electric motor")


# Creating objects
car1 = Car()
car2 = ElectricCar()

car1.drive()
car2.drive()
```

```python
### Polymorphism Using Common Interface

class Dog:
    def sound(self):
        print("Dog barks")

class Cat:
    def sound(self):
        print("Cat meows")

class Cow:
    def sound(self):
        print("Cow moos")


animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()
```

```python
class Payment:
    def pay(self, amount):
        pass

class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")

class UPI(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using UPI")

def process_payment(payment_method, amount):
    payment_method.pay(amount)


process_payment(CreditCard(), 1000)
process_payment(UPI(), 500)
```
