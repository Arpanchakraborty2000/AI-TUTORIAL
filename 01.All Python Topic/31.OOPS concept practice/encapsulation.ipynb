```python
#Public Members

#ðŸ‘‰ Accessible from anywhere
#ðŸ‘‰ Default in Python



class Person:
    def __init__(self, name, age):
        self.name = name      # Public variable
        self.age = age        # Public variable

    def greet(self):          # Public method
        print(f"Hello, my name is {self.name}")


p1 = Person("Arpan", 25)

# Accessible directly
print(p1.name)
print(p1.age)
p1.greet()
```

```python
#Protected Members

#ðŸ‘‰ Single underscore _variable
#ðŸ‘‰ Meant for internal use
#ðŸ‘‰ Accessible but NOT recommended outside class


class Employee:
    def __init__(self, name, salary):
        self.name = name
        self._salary = salary   # Protected variable

    def _show_salary(self):     # Protected method
        print(f"Salary: {self._salary}")


emp = Employee("Arpan", 50000)

# Technically accessible
print(emp._salary)
emp._show_salary()
```

```python
#Private Members

#ðŸ‘‰ Double underscore __variable
#ðŸ‘‰ Stronger restriction (name mangling)

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance   # Private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance


account = BankAccount("Arpan", 1000)

account.deposit(500)

print(account.get_balance())

# âŒ This will cause error
# print(account.__balance)
```

```python
#What are Getter and Setter?

#Getter â†’ Used to get (read) private variable

#Setter â†’ Used to set (update) private variable with validation

#Used mainly with private variables (__var)



class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance   # Private variable

    # Getter method
    def get_balance(self):
        return self.__balance

    # Setter method
    def set_balance(self, amount):
        if amount >= 0:
            self.__balance = amount
        else:
            print("Balance cannot be negative")


account = BankAccount("Arpan", 1000)

# Using getter
print(account.get_balance())

# Using setter
account.set_balance(2000)
print(account.get_balance())

# Invalid update
account.set_balance(-500)
```
