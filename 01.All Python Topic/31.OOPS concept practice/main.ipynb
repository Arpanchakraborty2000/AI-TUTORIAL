```python
#### Define class bank account

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
        
    def deposite(self,amaount):
        self.balance+=amaount
        print(f"{amaount} is deposited . New balance is {self.balance}")
        
    def withdrawl(self,amaount):
        if amaount>self.balance:
            print("f{amaount} is not present in this bank accaount ")
        else:
            self.balance-=amaount
            print(f"{amaount} is withdrawl . New balance is {self.balance}")
            
    def getbalance(self):
        return self.balance
    
    
bank1=BankAccount("Arpan",1000)
print(bank1.balance)
print(bank1.owner)
            
        
```
