```python
### inheritance (singel_inheritance)
## Parent class

class car:
    def __init__(self,window,doors,engine_type):
        self.window=window
        self.doors=doors
        self.engine_type=engine_type
        
    def drive(self):
        print(f"the person will drive the {self.engine_type} car")
        
        
```

```python
car1 = car(4,5,"petrol")
car1.drive()
```

```python
class tesla(car):
    def __init__(self, window, doors, engine_type,is_selfdrive):
        super().__init__(window, doors, engine_type)
        self.is_selfdrive=is_selfdrive
        
        
    def selfdrive(self):
        print(f"tesla support self driving : {self.is_selfdrive}")
```

```python

tesla1=tesla(4,5,"Eletric",True)
tesla1.selfdrive()

tesla1.drive()
```

```python
### Multiple Inheritance 
###  Base class 1

class Animal:
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        print("subclasss must be implemented in this method .... ")
        
###  Base class 2       
class pet:
    def __init__(self,owner):
        self.owner=owner
        
        
## Derive class

class Dog(Animal,pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        pet.__init__(self,owner)
        
    def speak(self):
        return f"{self.name} say woof"
    
    
    
## create object 

dog =Dog("doge","Arpan")
##dog.speak()
print(dog.owner)
print(dog.speak())
```
