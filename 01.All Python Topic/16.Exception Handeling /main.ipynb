```python
try:
    file = open('example.txt', 'r')
    content = file.read()
    print(content)

except FileNotFoundError:
    print("File does not exist...")

finally:
    if 'file' in locals() and not file.closed:
        file.close()
        print("File closed")
```

```python
### ## Exception hadelling with try , catch block and else , Finally

try:
    num= int(input("plz enter your number ...."))
    result = 10/num
    
except ValueError:
    print("this is not a valid number ..")
except ZeroDivisionError:
    print("plzz enter the number that is geater than 0 ..")
except Exception as ex:
    print(ex)
    
else:   ### If the exception is not occur the the else block is executed ...
    print(f"The result is {result}")
    
finally:   ### this finally block is executed all time wheather a exception error happend or not ....
    print("Execution is completed ...")
```
