```python
class vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y
        
    def __add__(self,others):
        return vector(self.x+others.x,self.y+others.y)
    def __sub__(self,others):
        return vector(self.x-others.x,self.y-others.y)
    def __mul__(self,others):
        return vector(self.x*others.x,self.y*others.y)
```
