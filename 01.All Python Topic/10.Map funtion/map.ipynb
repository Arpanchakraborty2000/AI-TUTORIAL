```python
def square(x):
    return x*x

square(10)
```

```python
number = [1,2,3,4,5,6,7,8]

(list(map(square,number)))
```

```python
## lamda funtion with map
number = [1,2,3,4,5,6,7,8]

list(map(lambda x: x*x ,number))
```

```python
## can we map multiple iterable 
number1= [1,2,3]
number2= [5,6,7]

add_number= map(lambda x,y: x+y,number1,number2)
print(list(add_number))
```

```python
### map to covert list of string to int 

## use map to typecating str -> int 

str_num = ['1','2','3','4']
int_num= map(int,str_num)
print(list(int_num))
```

```python
words= ['apple','banana','cherry']
upper_word= list(map(str.upper,words))

print(upper_word)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name':'Arpan','age':25},
    {'name':'jack','age':30}
]


list(map(get_name,people))
```
