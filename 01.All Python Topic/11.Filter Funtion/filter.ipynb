```python
def even(x):
    if x%2 ==0:
        return x
lst=[1,2,3,4,5,6,7,8]
list(filter(even,lst))
```

```python
## filter with lamda funtion

numbers =[1,2,3,4,5,6,7,8]
geter_then_five= list(filter(lambda x: x>5,numbers))

print(geter_then_five)
```

```python
## filter with lamda funtion with multiple condition

numbers =[1,2,3,4,5,6,7,8]

geter_then_five_and_even = list(filter(lambda x: x>5  and x%2==0,numbers))

print(geter_then_five_and_even)
```

```python
## pply finter to check age is gater then 25 in dictionary
people = [
    {'name':'Arpan','age':25},
    {'name':'jack','age':30},
    {'name':'john','age':35}
]

def age_is_gater_25(x):
    return x['age']>25

list(filter(age_is_gater_25,people))

```
