```python
## c;ass with instance method 
class Dog:
    def __init__(self,name,age):
        self.name=name  ## declaring the atribute 
        self.age=age
        
    ### creating the atribute 
    
    def bark(self):
       # print((self.name),"says woolf ")
        print(f"{self.name} says woolf")
        
dog1 = Dog("buddy",3)
dog1.bark()
dog2= Dog("Lucy",5)
dog2.bark()
```

```python
### Moideling a bank account 

## define class fo bank account 

class BankAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance= balance
        
    def deposite(self,amount):
        self.balance+=amount
        print(f"{amount} is deposited .New balance is {self.balance}")
        
    def withdrowl(self,amount):
        if amount>self.balance:
            print("Insufficient balance ...")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawl .New balance is {self.balance}")
            
    def get_balance(self):
        return self.balance
    
#### create a object 
acc1 = BankAccount("Arpan",5000)
print(acc1.balance)


## call instance method 

acc1.deposite(500)
acc1.withdrowl(300)
acc1.balance
```
