```python
def simple_gen():
    yield 1
    yield 2
    yield 3

gen = simple_gen()

print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3
# print(next(gen))  # ❌ Raises StopIteration
```

```python
## Practical example 
## reading large file ....

def read_large_file (file_path):
    with open(file_path,'r')as file:
        for line in file:
            yield line
    
file_path='large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
