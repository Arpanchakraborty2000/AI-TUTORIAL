```python
## Adding Multiplr line

lines=["First line\n,Second line\n,Third line\n"]
with open ('example.txt','a')as file:
    file.writelines(lines)
```

```python
## write a file without overwritting 

with open ('example.txt','a')as file:
    file.write("Adding line with no delete of existing line \n")
```

```python
## Writting the file with OVERWRITTEN 

with open ('example.txt','w')as file:
    file.write("Hello world\n")
    file.write("Arpan Chakraborty\n")
    file.write("I am adding a new line\n")

```

```python
## read line by line 
with open ('example.txt','r')as file:
    for line in file:
        print(line.strip()) ## Strip will remove the extra space for one line tyo another line ....
```

```python
## Read Full folder
with open ('example.txt','r')as file:
    content=file.read()
    print(content)
```
