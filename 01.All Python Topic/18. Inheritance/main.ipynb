```python
class Car:
    def __init__(self,windows, doors ,enginetype):
        
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
        
    def drive(self):
        print(f"The person will drive the {self.enginetype} car")
        
car1 =Car(4,5,"petrol")
print(car1.drive())


## Inheritance from parent class 


### child class inheriting the parent class ...

class Tesla(Car):
    def __init__(self, windows, doors, enginetype, is_selfdriving):
        super().__init__(windows, doors, enginetype)
        self.is_selfdriving=is_selfdriving
        
    def selfdriving(self):
        print(f"Tesla support self driving : {self.is_selfdriving}")
        
        
tesla1 = Tesla(4,5,"eletric","True")
tesla1.selfdriving()
tesla1.drive()
```

```python
### Multiple inheritance 
#### When one class inherits more then one base class ...

## Base class -> 1
class Animal():
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        print("Subclass must inplement this method")
        
## Base class -> 2

class Pet():
    def __init__(self,owner):
        self.owner=owner
        
### Derive class (Like implementing both class )

class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)  # Call Animal’s constructor
        Pet.__init__(self, owner)    # Call Pet’s constructor
        
    def speak(self):
        return f"{self.name} says woof"

    
    
## creating a object 

dog1 = Dog("Buddy", "Arpan")
print(dog1.speak())
print(f"Owner: {dog1.owner}")
```
